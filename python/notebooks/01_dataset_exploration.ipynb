```python
# Loading of the dataset files.

import pandas as pd
from pathlib import Path

base_path = Path("../../data/raw/msk_impact")

patients = pd.read_csv(base_path / "data_clinical_patient.txt", sep="\t", comment="#", low_memory=False)
samples = pd.read_csv(base_path / "data_clinical_sample.txt", sep="\t", comment="#", low_memory=False)
mutations = pd.read_csv(base_path / "data_mutations.txt", sep="\t", comment="#", low_memory=False)
```
